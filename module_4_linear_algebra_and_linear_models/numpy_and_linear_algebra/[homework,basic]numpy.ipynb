```python
def no_numpy_mult(a, b):
    n,m = len(a), len(a[0])
    k,l = len(b), len(b[0])
    
    result = [[] for i in range(n)]
    
    
    if m != k:
        return None
    
    for i in range(n):
        for r in range(l):
            row_v = []
            for j in range(k):
                row_v.append(b[j][r])
            
            result[i].append(sum(map(lambda x,y : x*y, a[i], row_v)))
        
    
    return result
```

```python
def numpy_mult(a, b):
    return np.dot(a,b)
```

### Задание 3
Напишите функцию, которая находим сумму четных диагональныех элементов квадратной матрицы. Если таких элементов нет, то вывести `0`. Используйте библиотеку Numpy

```python
# функция, решающая задачу с помощью NumPy
def diag_2k(a):
    sum_ = 0
    diag = np.diag(a)
    
    sum_ += diag[diag % 2 == 0].sum()
    return sum_
```

Написать функцию для кодирование массива (Run-length encoding). Все подряд повторения элементов функция сжимает в один элемент и считает количество повторений. Функция возвращает кортеж из двух векторов одинаковой длины. Первый содержит элементы, а второй — сколько раз их нужно повторить. 

Пример: encode(np.array([1, 2, 2, 3, 3, 1, 1, 5, 5, 2, 3, 3])) = (np.array[1, 2, 3, 1, 5, 2, 3]), np.array[1, 2, 2, 2, 2, 1, 2])

```python
# функция, решающая задачу с помощью NumPy
def encode(a):
    
    diff_arr = np.diff(a)
    diff_arr = np.insert(diff_arr, 1, 1)
    
    ind = np.where(diff_arr != 0)[0]
    
    return (a[ind], [len(item) for item in np.split(a, ind)[1:]])
```
